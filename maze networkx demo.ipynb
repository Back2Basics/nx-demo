## This is a 640x640 maze that we are going to solve in 5 min.

```python
maze = im.open('maze.png','r')
maze = maze.convert('1')
maze = maze.convert('RGB')
```

##     I'm going to define some colors so I don't have to keep thinking in or typing tuples.

```python
white=(255,255,255)
black=(0,0,0)
blue=(0,0,255)

start=(639,0)
end=(1,639)
```

## The Graph will look like this output but with 640x640.

```python
G= nx.Graph()
for i in range(3):
    for j in range(3):
        G.add_node((i,j))
        if i >0:
            G.add_edge((i-1,j),(i,j))
        if j>0:
            G.add_edge((i,j),(i,j-1))
nx.draw(G, pos=nx.spring_layout(G))
```

```python
G = nx.Graph()
#make a grid
for i in range(640):
    for j in range(640):
        G.add_node((i,j))
        if i >0:
            G.add_edge((i-1,j),(i,j))
        if j>0:
            G.add_edge((i,j),(i,j-1))
```

## Now that I have the graph that maps to the pixels in the maze, I'm going to delete the white pixels, which are the walls.

```python
for i in range(640):
    for j in range(640):
        if maze.getpixel((i,j))==white:
            G.remove_node((i,j))
```

## There are algorithms in networkx in order to get from node to node.  There is more than just A*.

```python
for x in nx.algorithms.astar_path(G,start,end):
    maze.putpixel(x,blue)
len(nx.algorithms.astar_path(G,start,end))  
```
